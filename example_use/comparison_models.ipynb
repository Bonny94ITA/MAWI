```python
import spacy
from spacy import displacy
```

##### IT

```python
nlp = spacy.load('it_core_news_sm', exclude=["ner"])
ner_ita = spacy.load('it_nerIta_trf')

nlp.add_pipe("transformer", name="trf_ita", source=ner_ita, last=True)
nlp.add_pipe("ner", name="ner_ita", source=ner_ita, last=True)

sentence = "Torino ospita anche alberi esotici secolari come le sequoie costali (Sequoia sempervirens) dei parchi collinari di Villa Genero e Giacomo Leopardi, i numerosi esemplari di Noce del Caucaso (Pterocarya fraxinifolia) che costeggiano il Po e compaiono in parchi pubblici come i Giardini Cavour, il giardino Sambuy, i Giardini Reali Bassi (lungo Corso San Maurizio), il parco Millefonti. "

doc_ner = nlp(sentence)

displacy.render(doc_ner, style="ent", jupyter=True)
displacy.render(doc_ner, style="dep", jupyter=True)
```

```python
nlp_no_trf = spacy.load("it_core_news_lg")

doc_no_trf = nlp_no_trf(sentence)

displacy.render(doc_no_trf, style="ent", jupyter=True)
displacy.render(doc_no_trf, style="dep", jupyter=True)
```

##### EN

```python
nlp = spacy.load("en_core_web_trf")
sentence = "This neighbourhood hosts the significant architecture of Santuario di Maria Ausiliatrice (\"Maria Ausiliatrice Sanctuary\") in the homonymous square and behind the church stands San Pietro in Vincoli old cemetery"

doc = nlp(sentence)

displacy.render(doc, style="ent", jupyter=True)
displacy.render(doc, style="dep", jupyter=True)
```

```python
nlp_no_trf = spacy.load("en_core_web_lg")

doc_no_trf = nlp_no_trf(sentence)

displacy.render(doc_no_trf, style="ent", jupyter=True)
displacy.render(doc_no_trf, style="dep", jupyter=True)
```
