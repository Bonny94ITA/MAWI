```python
import spacy

from collections import Counter

nlp = spacy.load("en_core_web_trf")

path = f"../input/articles1/en/texts/Turin.txt"

file = open(path, "r", encoding='utf-8') 
text = file.read() 
file.close() 

doc = nlp(text)

ents = doc.ents

ents_gpe = [ent for ent in ents if ent.label_ == "GPE"]

ents_gpe = [ent.text for ent in ents_gpe]

count = Counter(ents_gpe)

count.most_common()
```
